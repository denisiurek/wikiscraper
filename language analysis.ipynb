```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordfreq import get_frequency_dict
from wikiscraper import get_scraper_tool, ConfigLoader, crawl_subpages
```

```python
def language_confidence_score(word_counts: pd.DataFrame, lang: str) -> int:
    score = 0
    total_words = sum(word_counts.values())
    freq_dict = get_frequency_dict(lang)
    for word, count in word_counts.items():
        score += count * freq_dict[word] / total_words
    return score
```
